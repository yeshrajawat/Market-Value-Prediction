```python

clean_result["Name"] = clean_result["Name"].str.replace("\n","")
clean_result["Name"] = clean_result["Name"].str.replace("ST","")
clean_result["Name"] = clean_result["Name"].str.replace("CM","")
clean_result["Name"] = clean_result["Name"].str.replace("CAM","")
clean_result["Name"] = clean_result["Name"].str.replace("CB","")
clean_result["Name"] = clean_result["Name"].str.replace("CDM","")
clean_result["Name"] = clean_result["Name"].str.replace("GK","")
clean_result["Name"] = clean_result["Name"].str.replace("LB","")
clean_result["Name"] = clean_result["Name"].str.replace("RB","")
clean_result["Name"] = clean_result["Name"].str.replace("RW","")
clean_result["Name"] = clean_result["Name"].str.replace("LW","")
clean_result["Name"] = clean_result["Name"].str.replace("CF","")
clean_result["Name"] = clean_result["Name"].str.replace("RF","")
clean_result["Name"] = clean_result["Name"].str.replace("LF","")
clean_result["Name"] = clean_result["Name"].str.replace("LWB","")
clean_result["Name"] = clean_result["Name"].str.replace("RWB","")
clean_result["Name"] = clean_result["Name"].str.replace("LM","")
clean_result["Name"] = clean_result["Name"].str.replace("RM","")
clean_result["Team"] = clean_result["Team"].str.replace(r"[0-9,~]","")
clean_result["Team"] = clean_result["Team"].str.replace("ON LOAN","")
clean_result["Team"] = clean_result["Team"].str.replace("Free","")



clean_result["Team"] = clean_result["Team"].str.replace("\n","")
clean_result["Name"] = clean_result["Name"].str.strip()
clean_result["Age"] = clean_result["Age"].str.strip()
clean_result["OVA"] = clean_result["OVA"].str.strip()
clean_result["POT"] = clean_result["POT"].str.strip()
clean_result["Team"] = clean_result["Team"].str.strip()
clean_result["POT"] = clean_result["POT"].str.strip()
clean_result["Height"] = clean_result["Height"].str.strip()
clean_result["Weight"] = clean_result["Weight"].str.strip()
clean_result["foot"] = clean_result["foot"].str.strip()
clean_result["BOV"] = clean_result["BOV"].str.strip()
clean_result["BP"] = clean_result["BP"].str.strip()
clean_result["Growth"] = clean_result["Growth"].str.strip()
clean_result["Value"] = clean_result["Value"].str.strip()
clean_result["Wage"] = clean_result["Wage"].str.strip()
clean_result["Release Clause"] = clean_result["Release Clause"].str.strip()
clean_result["Attacking"] = clean_result["Attacking"].str.strip()


clean_result["Value"] = clean_result["Value"].str.replace(r"[€,M,K]","")
clean_result["Wage"] = clean_result["Wage"].str.replace(r"[€,K,k,M]","")
clean_result["Release Clause"] = clean_result["Release Clause"].str.replace(r"[€,M,K,k]","")

clean_result["Height"] = clean_result["Height"].str.replace("cm","")
clean_result["Weight"] = clean_result["Weight"].str.replace("kg","")
clean_result["Attacking"] = clean_result["Attacking"].str.replace("]","")
for col in column_names:
    
    clean_result[col] = clean_result[col].str.strip()
    clean_result[col] = clean_result[col].str.replace("]","")
```
